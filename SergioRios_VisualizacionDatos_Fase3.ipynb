<h3> 1. Cargue y Limpieza de datos. </h3>

```python
# URL de los datos
url = "https://www.datos.gov.co/api/views/gt2j-8ykr/rows.csv?accessType=DOWNLOAD&bom=true&format=true"
data = pd.read_csv(url)
```

```python
# Convertir todos los valores a minúsculas para facilitar la comparación
data['Estado'] = data['Estado'].str.lower()

# Crear un diccionario para mapear los valores a los valores estándar
mapping = {'activo': 'Activo', 'inactivo': 'Inactivo', 'n/a': 'Otro'}
```

```python
df = data[['fecha reporte web', 'Fecha de notificación', 'Nombre del país', 'Nombre departamento', 'Nombre municipio','Edad', 
           'Unidad de medida de edad', 'Sexo', 'Tipo de contagio', 'Ubicación del caso', 'Estado', 'Recuperado', 
           'Fecha de inicio de síntomas', 'Fecha de diagnóstico', 'Fecha de recuperación', 'Tipo de recuperación', 
           'Fecha de muerte']]
df['Nombre del país'].fillna('COLOMBIA', inplace=True)
```

```python
#Convirtiendo los valores FECHA de tipo object a formato DATE.
columnas_fechas = ['fecha reporte web', 'Fecha de notificación', 'Fecha de inicio de síntomas',
                  'Fecha de diagnóstico', 'Fecha de recuperación', 'Fecha de muerte']

# Convertir las columnas a formato datetime
for columna in columnas_fechas:
    df[columna] = pd.to_datetime(df[columna], format='%Y-%m-%d %H:%M:%S')
```

<h2> Desarrollo o ruta metodológica: </h2>

<h3> 1. Obtén un gráfico apropiado para cada uno de los siguientes ítems: </h3>

```python
# 1.1 Contagiados en Colombia, por fecha.

# Filtrar los datos de Colombia
df_colombia = df[df['Nombre del país'] == 'COLOMBIA']

# Agrupar los datos por fecha y contar los casos
casos_por_fecha_colombia = df_colombia.groupby('Fecha de notificación').size()

# Crear el gráfico de líneas
plt.figure(figsize=(12, 4))
plt.plot(casos_por_fecha_colombia.index, casos_por_fecha_colombia.values)
plt.xlabel('Fecha')
plt.ylabel('Número de casos')
plt.title('Contagios por fecha en Colombia')
plt.grid(True)
plt.show()
```

```python
# 1.2 Contagiados en Antioquia, por fecha.
# Filtrar los datos de Antioquia
df_antioquia = df[df['Nombre departamento'] == 'ANTIOQUIA']

# Agrupar los datos por fecha y contar los casos
casos_por_fecha_antioquia = df_antioquia.groupby('Fecha de notificación').size()

# Crear el gráfico de barras
plt.figure(figsize=(12, 4))
plt.bar(casos_por_fecha_antioquia.index, casos_por_fecha_antioquia.values, color='green')
plt.xlabel('Fecha')
plt.ylabel('Número de casos')
plt.title('Contagios por fecha en Antioquia')
plt.xticks(rotation=45)
plt.show()
```

```python
# 1.3 Contagiados en Medellín, por fecha
df_medellin = df[df['Nombre municipio'] == 'MEDELLIN']

# Agrupar los datos por fecha y contar los casos
casos_por_fecha_medellin = df_medellin.groupby('Fecha de notificación').size()

# Crear el gráfico de barras
plt.figure(figsize=(12, 4))
plt.bar(casos_por_fecha_medellin.index, casos_por_fecha_medellin.values, color='red')
plt.xlabel('Fecha')
plt.ylabel('Número de casos')
plt.title('Contagios por fecha en Medellín')
plt.xticks(rotation=45)
plt.show()
```

```python
# 1.4 Muertes en Colombia, por fecha.

# Filtrar los datos de Colombia y los casos fallecidos
df_muertes_colombia = df[(df['Nombre del país'] == 'COLOMBIA') & (df['Recuperado'] == 'Fallecido')]

# Agrupar los datos por fecha de muerte y contar los casos
muertes_por_fecha_colombia = df_muertes_colombia.groupby('Fecha de muerte').size()

plt.figure(figsize=(12, 4))
plt.plot(muertes_por_fecha_colombia.index, muertes_por_fecha_colombia.values, color='blue')
plt.xlabel('Fecha de muerte')
plt.ylabel('Número de muertes')
plt.title('Muertes por COVID-19 en Colombia')
plt.grid(True)
plt.show()
```

```python
# 1.5 Muertes en Antioquia, por fecha.

# Filtrar los datos de Antioquia y los casos fallecidos
df_muertes_antioquia = df[(df['Nombre departamento'] == 'ANTIOQUIA') & (df['Recuperado'] == 'Fallecido')]

# Agrupar los datos por fecha de muerte y contar los casos
muertes_por_fecha_antioquia = df_muertes_antioquia.groupby('Fecha de muerte').size()

plt.figure(figsize=(12, 4))
plt.plot(muertes_por_fecha_antioquia.index, muertes_por_fecha_antioquia.values, color='green')
plt.xlabel('Fecha de muerte')
plt.ylabel('Número de muertes')
plt.title('Muertes por COVID-19 en Antioquia')
plt.grid(True)
plt.show()
```

```python
# 1.6 Muertes en Medellín, por fecha.

# Filtrar los datos de Antioquia y los casos fallecidos
df_muertes_medellin = df[(df['Nombre municipio'] == 'MEDELLIN') & (df['Recuperado'] == 'Fallecido')]

# Agrupar los datos por fecha de muerte y contar los casos
muertes_por_fecha_medellin = df_muertes_medellin.groupby('Fecha de muerte').size()

plt.figure(figsize=(12, 4))
plt.plot(muertes_por_fecha_medellin.index, muertes_por_fecha_medellin.values, color='red')
plt.xlabel('Fecha de muerte')
plt.ylabel('Número de muertes')
plt.title('Muertes por COVID-19 en Medellín')
plt.grid(True)
plt.show()
```

<h3> 2. Construye un pairplot. </h3>

```python
# Creación de nueva columna con valores Bool para registrar 1 si el pacienete fallecio y 0 si sobrevivió.
df['fallecido'] = np.where(df['Recuperado'] == 'Fallecido', 1, 0)

# Nueva columna para calcular los días que pasaron desde el inicio de sintomas hasta que fue atendido.
df['dias_para_atencion'] = (df['Fecha de diagnóstico'] - df['Fecha de inicio de síntomas']).dt.days

# Nueva columna para calcular los días para recuperación.
df['dias_recuperacion'] = (df['Fecha de recuperación'] - df['Fecha de inicio de síntomas']).dt.days
```

```python
# Seleccionar solo las variables numéricas adecuadas
df_numerico = df[['Edad', 'dias_para_atencion', 'dias_recuperacion']]

plt.figure(figsize=(6, 4))

# Crear el pairplot
sns.pairplot(df_numerico)
plt.show()
```
